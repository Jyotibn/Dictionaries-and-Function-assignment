```python
class area:
    def __init__(self,side1,side2,side3):
        self.a=side1
        self.b=side2
        self.c=side3
    
class area_of_triangle(area):
   
    def area_t(self):
        self.s=(self.a+self.b+self.c)/2
        self.A=(self.s*(self.s-self.a)*(self.s-self.b)*(self.s-self.c))**0.5
        return self.A
   
```

```python
t=area_of_triangle(4,5,6)
print("area of triangle=",t.area_t())

```

```python
def filter_long_words(l,n):
    l1=[]
    for i in l:
        if len(i)>n:
            l1.append(i)
    return l1
    
```

```python
filter_long_words(["jyoti","malik","abcde","ro","k","happiness"],4)
```

```python
def length(l):
    l1=[]
    for i in l:
        l1.append(len(i))
    return l1
```

```python
length(["ab","cde","erty"])
```

```python
def vowel(str):
    if len(str)==1:
        if (str=='a')or(str=='e')or(str=='i')or(str=='o')or(str=='u'):
            return True
        else:
            return False
    else:
        return "enter a string of length 1"
    
    
```
